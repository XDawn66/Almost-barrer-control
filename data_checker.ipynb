```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


file_path = "data/safe_data_set.pkl"  # Adjust to your actual file path
df = pd.read_pickle(file_path)
```

```python
df = pd.DataFrame(df, columns=["X", "Y", "Angle", "Speed"])
df.shape
```

```python
df.describe()
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df["X"], df["Y"], c=df["Speed"], cmap="viridis", alpha=0.7)
plt.colorbar(scatter, label="Speed")

plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("State Distribution (Color by Speed)")
plt.grid(True)
plt.show()
```
